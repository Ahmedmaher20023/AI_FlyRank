```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 1. Traffic skew — justifies clustering over a flat ranking/average
sorted_sessions = df["sessions_90d"].sort_values(ascending=False)
top10pct_n = int(len(df) * 0.10)
top10pct_share = sorted_sessions.head(top10pct_n).sum() / sorted_sessions.sum()

# 2. Zero-engagement share — shows a large, heterogeneous "silent" segment
zero_engagement_share = (df["engaged_sessions_90d"] == 0).mean()

# 3. Word count spread — shows a second independent axis pages vary on
wc_spread = df["word_count"].quantile([0.1, 0.9]).to_dict()

print(f"Top 10% of pages account for {top10pct_share:.1%} of all sessions_90d")
print(f"{zero_engagement_share:.1%} of pages have zero engaged_sessions_90d")
print(f"Word count spread (p10 to p90): {wc_spread[0.1]:.0f} to {wc_spread[0.9]:.0f} words")
```
