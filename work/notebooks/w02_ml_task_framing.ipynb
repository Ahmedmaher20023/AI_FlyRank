## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
#Rigid cutoffs fail: An if statement creates arbitrary cliffs (e.g., failing a piece of content at 9.9% CTR when the rule is 10%). ML groups items naturally by their overall behavioral proximity.

#Too many variables: Metrics interact. A 40% completion rate is great for a long deep-dive, but terrible for a short clip. Writing nested rules for every combination is impossible; ML calculates these multi-dimensional interactions automatically.

#Blind spots: Fixed rules force data into assumptions you already have. Unsupervised ML discovers hidden, emerging audience behaviors you wouldn't even know to write a rule for.
```
